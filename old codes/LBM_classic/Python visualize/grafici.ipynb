```python
import matplotlib.pyplot as plt
```

```python
x=[100,200,500,1000,2000,5000]
y=[242,458,1880,8841,53388,509222]

# Scatter plot
plt.figure(figsize=(8, 6))
#plt.scatter(x, y, color='blue', edgecolor='black', s=100, alpha=0.7)
plt.plot(x, y, marker='o', linestyle='-', color='b')
t=[100,500,1000,2000,5000]
plt.xticks(t)
plt.title('Scalability', fontsize=16)
plt.xlabel('Grid Dimension', fontsize=14)
plt.ylabel('Computation Time (ms)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)


# Show the plot
plt.savefig("Scalability_1.png", dpi=300)
plt.show()
plt.close()
```
